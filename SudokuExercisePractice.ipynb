```python
def is_valid_move(board, row, col, num):
    """Check if placing `num` at (row, col) is valid."""
    if num in board[row]:
        return False
    if num in [board[i][col] for i in range(9)]:
        return False
    box_row, box_col = 3 * (row // 3), 3 * (col // 3)
    for i in range(box_row, box_row + 3):
        for j in range(box_col, box_col + 3):
            if board[i][j] == num:
                return False
    return True

def is_valid_board(board):
    """Validate the entire board for correctness (no duplicates in rows/cols/boxes)."""
    def no_duplicates(lst):
        nums = [x for x in lst if x != 0]
        return len(nums) == len(set(nums))

    for i in range(9):
        if not no_duplicates(board[i]):
            return False
        if not no_duplicates([board[r][i] for r in range(9)]):
            return False

    for br in range(3):
        for bc in range(3):
            box = [
                board[br * 3 + r][bc * 3 + c]
                for r in range(3) for c in range(3)
            ]
            if not no_duplicates(box):
                return False
    return True

def is_solved(board):
    """Check if the board is completely and correctly solved."""
    return all(board[i][j] != 0 for i in range(9) for j in range(9)) and is_valid_board(board)
```

```python
def is_valid_move(board, row, col, num):
    """Check if placing `num` at (row, col) is valid."""
    if num in board[row]:
        return False
    if num in [board[i][col] for i in range(9)]:
        return False
    box_row, box_col = 3 * (row // 3), 3 * (col // 3)
    for i in range(box_row, box_row + 3):
        for j in range(box_col, box_col + 3):
            if board[i][j] == num:
                return False
    return True

def is_valid_board(board):
    """Validate the entire board for correctness (no duplicates in rows/cols/boxes)."""
    def no_duplicates(lst):
        nums = [x for x in lst if x != 0]
        return len(nums) == len(set(nums))

    for i in range(9):
        if not no_duplicates(board[i]):
            return False
        if not no_duplicates([board[r][i] for r in range(9)]):
            return False

    for br in range(3):
        for bc in range(3):
            box = [
                board[br * 3 + r][bc * 3 + c]
                for r in range(3) for c in range(3)
            ]
            if not no_duplicates(box):
                return False
    return True

def is_solved(board):
    """Check if the board is completely and correctly solved."""
    return all(board[i][j] != 0 for i in range(9) for j in range(9)) and is_valid_board(board)
```
