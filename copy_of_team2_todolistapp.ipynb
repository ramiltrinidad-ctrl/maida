```python
import pandas as pd
import io
from google.colab import files
```

```python
def display_menu():
    print("\n--- TO-DO LIST MENU ---")
    print("1. View Tasks")
    print("2. Add Task")
    print("3. Remove Specific Task")
    print("4. Upload from Excel")
    print("5. Delete ALL Tasks")
    print("6. STOP APPLICATION") # The Stop Option
    print("-----------------------")
```

```python
def view_tasks(todo_list):
    if not todo_list:
        print("\nYour list is empty!")
    else:
        print("\nYOUR TASKS:")
        for index, task in enumerate(todo_list, start=1):
            print(f"{index}. {task}")
```

```python
def add_task(todo_list):
    task = input("Enter the task: ")
    todo_list.append(task)
    print("Task added!")
```

```python
def upload_tasks(todo_list):
    print("\nPlease upload an Excel file (.xlsx) with a column header named 'Task'.")
    try:
        uploaded = files.upload()
        for filename in uploaded.keys():
            df = pd.read_excel(io.BytesIO(uploaded[filename]))
            if 'Task' in df.columns:
                new_tasks = df['Task'].dropna().astype(str).tolist()
                todo_list.extend(new_tasks)
                print(f"Successfully added {len(new_tasks)} tasks!")
            else:
                print("Error: Could not find 'Task' column.")
    except Exception as e:
        print(f"Upload cancelled or failed: {e}")
```

```python
def delete_all_tasks(todo_list):
    confirm = input("Are you sure you want to delete EVERYTHING? (y/n): ").lower()
    if confirm == 'y':
        todo_list.clear()
        print("All tasks deleted.")
    else:
        print("Action cancelled.")
```

```python
def main():
    tasks = []
    print("Welcome to Team 2's To-Do Manager!")

    try:
        while True:
            display_menu()
            choice = input("\nChoose an option (1-6): ")

            if choice == '1':
                view_tasks(tasks)
            elif choice == '2':
                add_task(tasks)
            elif choice == '3':
                # Simplified removal logic
                view_tasks(tasks)
                if tasks:
                    try:
                        num = int(input("Enter number to remove: "))
                        tasks.pop(num - 1)
                        print("Task removed.")
                    except:
                        print("Invalid input.")
            elif choice == '4':
                upload_tasks(tasks)
            elif choice == '5':
                delete_all_tasks(tasks)
            elif choice == '6':
                print("\nShutting down... Goodbye!")
                break # This stops the loop and the program
            else:
                print("Invalid choice, try again.")

    except KeyboardInterrupt:
        # This triggers if you click the 'Stop' button in Google Colab
        print("\n\nApplication stopped manually. Data lost unless saved.")

# Run the app
main()
```
