Autograded answers are usually followed by one or more cells containing automated tests.  
The visible tests are the `assert` statements.  
You **cannot** delete or edit them.  
**Do not** copy and paste these cells.

Note that the visible tests exist both as part of the grading, and to clarify ambiguities in the intended behavior.  

There can also be hidden tests.  
Below I show one such hidden test that would be part of the grading, but would not be shown in the actual assignment.  

```python
from numbers import Number
from numpy.testing import assert_almost_equal

area = area_2x2()
assert isinstance(area, Number)
### BEGIN HIDDEN TESTS
assert_almost_equal(area, 4)
### END HIDDEN TESTS
```
